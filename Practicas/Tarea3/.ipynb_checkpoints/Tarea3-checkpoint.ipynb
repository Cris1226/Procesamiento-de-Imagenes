### c) Implementa el algoritmo de especificación de histograma de una imagen en escala de grises, cuyo procedimiento se resume en la figura 3.


````python
def especificar_histograma(img_origen, hist_referencia):
    
    # Calcula el histograma de la imagen de origen
    hist_origen = cv2.calcHist([img_origen], [0], None, [256], [0, 256])
    
    # Calcula las funciones de distribución acumulativa (CDF)
    cdf_origen = hist_origen.cumsum()
    cdf_referencia = hist_referencia.cumsum()
    
    # Mapea los valores de CDF de la imagen de origen a los valores de CDF de la imagen de referencia
    mapeo = np.interp(cdf_origen, cdf_referencia, np.arange(256))
    
    # Aplica la transformación a la imagen de origen
    img_resultante = mapeo[img_origen]
    
    # Convierte la imagen resultante a tipo uint8 (0-255)
    img_resultante = np.uint8(img_resultante)
    
    return img_resultante

# Cargar la imagen de origen que se desea modificar
img_origen = cv2.imread('Imagenes/emma.jpg', cv2.IMREAD_GRAYSCALE)

# Cargar la imagen de referencia que proporciona el histograma deseado
img_referencia = cv2.imread('Imagenes/CieloNoc.jpg', cv2.IMREAD_GRAYSCALE)

# Calcula el histograma de la imagen de referencia
hist_referencia = cv2.calcHist([img_referencia], [0], None, [256], [0, 256])

# Aplica la especificación de histograma
img_resultante = especificar_histograma(img_origen, hist_referencia)

# Muestra la imagen original, la imagen de referencia y la imagen resultante
cv2.imshow('Origen', img_origen)
cv2.imshow('Referencia', img_referencia)
cv2.imshow('Resultante', img_resultante)

cv2.waitKey(0)
cv2.destroyAllWindows()

````

## a) Averigüe como crear un histograma en 2d con OpenCV.

Guiándonos de la documentación de OpenCV pudimos comprender como crear histogramas en 2 dimensiones.

A diferencia de los histogramas que hemos visto en el curso en los histogramas 2D, se consideran dos características a la vez. Esto es útil para encontrar histogramas de color donde las dos características son los valores de matiz (Hue) y saturación (Saturation) de cada píxel.

### Convertir al espacio HSV: 
Por esto, el primer paso para utilizar estos histogramas es pasar nuestra imagen al espectro HSV con la función cv.cvtColor(). De la siguiente forma:

```sh
hsv = cv.cvtColor(img, cv.COLOR_BGR2HSV)
```
### Cálculo del histograma:
A partir de lo anterior, utilizaremos los 2 primeros canales de la imagen referentes al matiz y la saturación para calcula el histograma 2D. Para este fin se hace uso de la función cv.calcHist() ya vista en clase, sin embargo, es importante especificar los siguientes argumentos:

    - El espacio de color HSV como imagen de entrada.
    - Los canales para el cálculo del histograma, que son el canal 0 (matiz) y el canal 1 (saturación).
    - La máscara, que se establece en None para incluir todos los píxeles.
    - El número de bins (divisiones) para el matiz y la saturación, que son 180 bins para el matiz y 256 bins para la saturación.
    - El rango de valores para el matiz y la saturación, que varían de 0 a 180 y de 0 a 256, respectivamente.
    
La función quedaría estructurada de la siguiente forma: 
```sh
hist = cv.calcHist([hsv], [0, 1], None, [180, 256], [0, 180, 0, 256])
```

### Ploteo del histograma:
Para visualizar el histograma 2D, se puede utilizar la función plt.imshow() de Matplotlib. Es importante especificar interpolation='nearest' para obtener una representación nítida del histograma.

```sh
plt.imshow(hist, interpolation='nearest')
plt.show()
```

````python
img = cv.imread('Imagenes/CieloNoc.jpg')

hsv = cv.cvtColor(img,cv.COLOR_BGR2HSV)
hist = cv.calcHist( [hsv], [0, 1], None, [180, 256], [0, 180, 0, 256] )
plt.imshow(hist,interpolation = 'nearest')
plt.show()
````

##  b) Defina una función que ingresando un rango de entrada (ej. 100, 120), se cree una función para expandir ese rango, en todo el rango dinámico de 0 a 255

````python
def expandir_rango(rango_entrada, rango_dinamico=(0, 255)):
    # Obtener los límites del rango de entrada
    min_entrada, max_entrada = rango_entrada

    # Obtener los límites del rango dinámico (0 a 255)
    min_dinamico, max_dinamico = rango_dinamico

    # Calcular la pendiente y la ordenada al origen de la transformación lineal
    pendiente = (max_dinamico - min_dinamico) / (max_entrada - min_entrada)
    ordenada_origen = min_dinamico - pendiente * min_entrada

    # Aplicar la transformación lineal a cada valor en el rango de entrada
    rango_expandido = np.arange(min_entrada, max_entrada + 1)
    rango_expandido = (rango_expandido * pendiente + ordenada_origen).astype(int)

    return rango_expandido

# Ejemplo de uso
rango_entrada = (100, 150)
valores_expandidos = expandir_rango(rango_entrada)

print(valores_expandidos)
````

### c) Definir una función para aumentar o reducir el brillo de una imagen, según valor especificado.

````python
def ajustar_brillo(img, valor):

    img_ajustada = np.clip(img.astype(int) + valor, 0, 255).astype(np.uint8)
    return img_ajustada

# Cargar una imagen de ejemplo
img = cv2.imread('Imagenes/emma.jpg')

# Ajustar el brillo
valor = int(input("Ingrese el valor de brillo a modificar (de -100 hasta 100): "))
valor = valor*255/100

img_ajustada = ajustar_brillo(img, valor)

# Mostrar la imagen original y la imagen ajustada
cv2.imshow('Imagen Original', img)
cv2.imshow('Imagen Ajustada', img_ajustada)
cv2.waitKey(0)
cv2.destroyAllWindows()
````
